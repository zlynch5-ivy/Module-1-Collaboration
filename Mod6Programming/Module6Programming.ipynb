```python
import time
fmt = "%Y %m %d"
notepad = open('today.txt', 'rt')
today_string = notepad.read()
theDay = time.strptime(today_string, fmt)
print(theDay)
notepad.close()
```

```python
import multiprocessing
import random
import time
from datetime import datetime

def worker():
    pause = random.random()
    time.sleep(pause)
    print_time = datetime.now().strftime("%H:%M:%S.%f")
    print(f"Process waited {pause:.4f} second(s). Current time: {print_time}.")

if __name__ == "__main__":
    processes = []

    for i in range(3):
        p = multiprocessing.Process(worker())
    
    print("Processes finished.")
```
