```python
# Zach Lynch

# Week2Lab

# App takes names and GPA, and compares the latter with a set value.
# Then the app tells the user whether the given student has made
# Honor Roll, Dean's List, or neither - with 'ZZZ' being input
# to break from the loop.


last_name = ""

while last_name != "ZZZ":
    last_name = input("Please input a student's last name. Type 'ZZZ' to quit.")

    if last_name == "ZZZ":
        break

    first_name = input("Please enter a student's first name.")
    gpa = float(input("Please enter a student's GPA."))

    if gpa >= 3.5:
        print(first_name, last_name,"has made the Dean's List.")

    elif gpa >= 3.25:
        print(first_name, last_name,"has made the Honor Roll.")

    else:
        print(first_name, last_name,"has not made the Honor Roll or Dean's List.")
```
